```python
import pandas as pd
import matplotlib.pyplot as plt

# Lendo o csv
df = pd.read_csv('O que te motiva a doar para ONGs_.csv')
df.head()
```

```python
colunas_perguntas = [
    'De 1 a 5, o quanto se identificar com uma causa social te motiva a doar?',
    'De 1 a 5, o quanto sua religião/devoção te motiva a ser solidário com uma causa?',
    'De 1 a 5, o quanto estar próximo ou acompanhar pessoas realizando doações te incentiva a ter atividade filantrópicas? (Filantropia: ato de ajudar o próximo, por meio de várias atitudes altruístas e solidárias).',
    'De 1 a 5, o quanto já é um hábito em sua família/grupo de realizar doações?',
    'De 1 a 5, o quanto a transparência em uma organização social de motiva a contribuir com ela?',
    'De 1 a 5, o quanto acompanhar as ações realizadas por um instituto te faz doar para ele?',
    'De 1 a 5, o quanto ter uma forma fácil e segura de realizar as doações te incentiva a doar?',
    'De 1 a 5, quanto o sentimento de compaixão te leva a doar?',
    'De 1 a 5, com qual frequência você realiza doações?'
]

# Calculando a media das perguntas
medias = df[colunas_perguntas].mean()

medias.index = [f'pgt{i+1}' for i in range(len(medias))]

# Plotar um gráfico da media de todas as perguntas
plt.figure(figsize=(10, 5))
medias.plot(kind='bar', color='lightblue')
plt.title('Motivos que levam as pessoas a doarem')
plt.xlabel('Perguntas')
plt.ylabel('Média')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Plotar grafico das perguntas isoladamente
for i, coluna in enumerate(colunas_perguntas, start=1):
    plt.figure(figsize=(6, 4))
    plt.hist(df[coluna], bins=5, rwidth=0.9, color=f'C{i}')
    plt.title(f'Pergunta {i}')
    plt.xlabel('Valor')
    plt.ylabel('Pessoas')
    plt.show()
```
